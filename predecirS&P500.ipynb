```python
import yfinance as yf
sp500 = yf.Ticker("^GSPC")
sp500 = sp500.history(period="max")
sp500
```

```python
sp500 = sp500.drop(["Dividends", "Stock Splits"], axis = 1)
sp500
```

```python
sp500["Precio maÃ±ana"] = sp500["Close"].shift(-1)
sp500["Alza"] = (sp500["Precio maÃ±ana"] > sp500["Close"]).astype(int)
sp500 = sp500.loc["2014-01-01":].copy()
sp500 = sp500.dropna()
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score
from sklearn.model_selection import train_test_split

model = RandomForestClassifier(n_estimators=500, min_samples_split=250, random_state=1, max_depth=3, criterion="entropy")

X = sp500[[ "Open", "High", "Low", "Close", "Volume" ]]
y = sp500[['Alza']].values.ravel()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42, shuffle=True)
model.fit(X_train, y_train)
preds = model.predict(X_test)
precision = precision_score(y_test, preds)
print("PrecisiÃ³n:", precision)
```
